## HEATMAP
The correlation heatmap further confirmed these relationships statistically. The Survived variable showed a positive correlation with Fare (0.26) and a negative correlation with Pclass (-0.34), meaning passengers who paid higher fares and were from a higher class were more likely to survive. There was also a strong negative correlation between Pclass and Fare (-0.55), suggesting that first-class passengers paid significantly more than those in third class. Other features, however, exhibited weaker correlations with survival.

```python
# Calculate the correlation matrix
corr_matrix = df[['Age', 'Fare', 'Pclass', 'SibSp', 'Parch']].corr()

# Enhanced Heatmap with annotations
plt.figure(figsize=(10, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', cbar=True, linewidths=0.5)
plt.title('Correlation Heatmap of Numerical Features', fontsize=16)
plt.show()
```

## Age distribution by survival status
The shapes of the distributions suggest that younger individuals had a higher risk of death, while an increasing proportion of older individuals survived, or fewer deaths were observed.

```python
plt.figure(figsize=(8, 6))
sns.histplot(df[df['Survived'] == 0]['Age'], kde=True, label='Died', color='red', bins=20, alpha=0.6)
sns.histplot(df[df['Survived'] == 1]['Age'], kde=True, label='Survived', color='green', bins=20, alpha=0.6)
plt.title('Age Distribution by Survival Status (Histogram)', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend()
plt.show()
```

## Bar plot for survival by gender
This plot shows the proportion of survivors (Survived = 1) in each gender group. Historically, more women survived due to the "women and children first" policy.

```python
plt.figure(figsize=(8, 6))
sns.barplot(x='Sex', y='Survived', data=df, palette='coolwarm')
plt.title('Survival Rate by Gender', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Survival Rate', fontsize=12)
plt.show()
```

## Violin Plot for Age Distribution by Passenger Class
The violin plot shows how the age distribution varies across 1st, 2nd, and 3rd class passengers. It also helps you understand whether there were age-related trends within different classes.

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x='Pclass', y='Age', data=df, palette='coolwarm')
plt.title('Age Distribution by Passenger Class', fontsize=16)
plt.xlabel('Passenger Class', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.show()
```

## Countplot for Survival by Embarked (Port of Embarkation)
This plot helps compare survival rates by the port from which passengers embarked.
You might notice that passengers who embarked from Cherbourg (C) had a higher survival rate compared to those from Southampton (S) or Queenstown (Q).

```python
# Full correct version

# 1. Clean column names if needed
df.columns = df.columns.str.strip()

# 2. Convert 'Survived' to string (already done)
df['Survived'] = df['Survived'].astype(str)

# 3. Plot using correct column names (capital 'Embarked')
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='Embarked', hue='Survived', palette='pastel')
plt.title('Survival Count by Port of Embarkation', fontsize=16)
plt.xlabel('Port of Embarkation (C = Cherbourg, Q = Queenstown, S = Southampton)', fontsize=12)
plt.ylabel('Number of Passengers', fontsize=12)
plt.legend(title='Survived', labels=['Did Not Survive', 'Survived'])
plt.show()
```

## Fare by Survival Status
Boxplots reveal if higher fares are associated with higher survival rates, which is typically the case. Wealthier passengers (1st class) might have had more access to lifeboats.
The plot shows the median, quartiles, and potential outliers in Fare for survivors and non-survivors.

```python
# Step 1: Fix column names if needed
df.columns = df.columns.str.strip()

# Step 2: Convert 'Survived' to string for better plotting
df['Survived'] = df['Survived'].astype(str)

# Step 3: Create a better-looking plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Survived', y='Fare', data=df, palette='coolwarm')

# Step 4: Titles and labels to tell the story
plt.title('Fare vs Survival Status', fontsize=18, fontweight='bold')
plt.xlabel('Survival Status (0 = Did Not Survive, 1 = Survived)', fontsize=14)
plt.ylabel('Fare Paid', fontsize=14)
plt.grid(True, axis='y', linestyle='--', alpha=0.7)

# Step 5: Show plot
plt.show()

```

## Survival by age group
Children had the highest survival rate, followed by teens and adults; middle-aged and seniors had lower survival chances overall.

```python
# Load Titanic dataset (Seaborn has it built-in)
titanic = sns.load_dataset('titanic')

# Create Age Groups
bins = [0, 12, 18, 35, 60, 100]
labels = ['Child (0-12)', 'Teen (13-18)', 'Adult (19-35)', 'Middle Age (36-60)', 'Senior (60+)']
titanic['age_group'] = pd.cut(titanic['age'], bins=bins, labels=labels)

# Calculate Survival Rate by Age Group
survival_rate = titanic.groupby('age_group')['survived'].mean().reset_index()

# Plotting
plt.figure(figsize=(8,6))
sns.barplot(x='age_group', y='survived', data=survival_rate, palette='coolwarm')

plt.title('Survival Rate by Age Group', fontsize=16)
plt.ylabel('Survival Rate', fontsize=14)
plt.xlabel('Age Group', fontsize=14)
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Family Size vs Survival (SibSp + Parch)
Survival rate tends to be higher for individuals with smaller family sizes (1-3). Larger family sizes have a lower survival rate.

```python
# Load Titanic dataset (Seaborn has it built-in)
titanic = sns.load_dataset('titanic')

# Calculate family size (SibSp + Parch)
titanic['family_size'] = titanic['sibsp'] + titanic['parch']

# Plotting Family Size vs Survival
plt.figure(figsize=(8,6))
sns.barplot(x='family_size', y='survived', data=titanic, palette='coolwarm')

plt.title('Survival Rate by Family Size', fontsize=16)
plt.ylabel('Survival Rate', fontsize=14)
plt.xlabel('Family Size (SibSp + Parch)', fontsize=14)
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## Survival by Deck (first letter of Cabin)
Passengers on higher decks (B, D, E) typically had a higher survival rate.

```python
# Load Titanic dataset (Seaborn has it built-in)
titanic = sns.load_dataset('titanic')

# Filter out missing values in the 'deck' column
titanic = titanic[titanic['deck'].notna()]

# Plotting Survival Rate by Deck
plt.figure(figsize=(8,6))
sns.barplot(x='deck', y='survived', data=titanic, palette='coolwarm')

plt.title('Survival Rate by Deck', fontsize=16)
plt.ylabel('Survival Rate', fontsize=14)
plt.xlabel('Deck', fontsize=14)
plt.ylim(0, 1)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
